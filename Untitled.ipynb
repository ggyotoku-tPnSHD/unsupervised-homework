```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
file = "Resources/crypto_data.csv"
df = pd.read_csv(file)
```

```python
currently_trading = df.loc[df["IsTrading"] == True, :]
currently_trading = currently_trading.dropna(axis=0)
currently_trading["TotalCoinSupply"] = currently_trading["TotalCoinSupply"].astype(
    "float"
)
currently_trading.info()
```

```python
currently_trading = currently_trading.loc[currently_trading["TotalCoinSupply"] > 0, :]
currently_trading = currently_trading.drop(
    columns=["CoinName", "Unnamed: 0", "IsTrading"]
)
currently_trading = pd.get_dummies(currently_trading)
```

```python
currently_trading
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
coin_scaled = scaler.fit_transform(currently_trading)
```

```python
from sklearn.decomposition import PCA

pca_test = PCA(n_components=0.9)

coin_pca = pca_test.fit_transform(coin_scaled)
coin_pca_df_test = pd.DataFrame(data=coin_pca)
pca_test.explained_variance_ratio_
```

```python
pca = PCA(n_components=3)

coin_pca = pca.fit_transform(coin_scaled)
coin_pca_df = pd.DataFrame(data=coin_pca)
# coin_pca_df
```

```python
# Finding the best value for k
inertia = []
k = list(range(1, 10))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(coin_pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow["k"], df_elbow["inertia"])
plt.xticks(list(range(10)))
plt.title("Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
model = KMeans(n_clusters=5, random_state=0)
model.fit(coin_pca_df)
predictions = model.predict(coin_pca_df)
coin_pca_df["class"] = model.labels_
coin_pca_df.columns
```

```python
# analysis = coin_pca_df.loc[coin_pca_df['class']==2]
```

```python
df = coin_pca_df.drop(["class"], axis=1)
labels = coin_pca_df["class"]
```

```python
tsne = TSNE(learning_rate=35)
tsne_feature = tsne.fit_transform(df)
df["x"] = tsne_feature[:, 0]
df["y"] = tsne_feature[:, 1]
```

## Conclusion
- The new coins entering the market are can be classifed into different clustered as shown in the graphs.

```python
plt.scatter(df["x"], df["y"], c=labels)
plt.show
```
